# Exercise 3

```python
# 1. Load Image and Convert to Grayscale
img = Image.open("/content/drive/MyDrive/AI ML/week1/images.png").convert("L")
img = np.array(img)

plt.imshow(img, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Standardize Data
mean = np.mean(img)
std = np.std(img)

standardized = (img - mean) / std
```

```python
# Covariance Matrix
cov_matrix = np.cov(standardized)
```

```python
# 2 Eigen Decomposition
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)
```

```python
# Sort Eigenvalues
idx = np.argsort(eigen_values)[::-1]

eigen_values = eigen_values[idx]
eigen_vectors = eigen_vectors[:,idx]
```

```python
# Explained Variance Plot
explained = eigen_values / np.sum(eigen_values)
cumulative = np.cumsum(explained)

plt.plot(cumulative)
plt.title("Cumulative Explained Variance")
plt.xlabel("Principal Components")
plt.ylabel("Variance")
plt.show()
```

```python
# Select Top k Components
k = 50
pcs = eigen_vectors[:,:k]
```

```python
# 3 Transform Image
compressed = np.dot(pcs.T, standardized)
```

```python
# Reconstruct Image
reconstructed = np.dot(pcs, compressed)
```

```python
# Display Reconstructed Image
import matplotlib.pyplot as plt

plt.imshow(reconstructed, cmap="gray")
plt.title("Reconstructed Image")
plt.axis("off")
plt.show()
```

```python
# Experiment with different PC values
import matplotlib.pyplot as plt
ks=[10,30,60,100]

for k in ks:
    pcs=eigen_vectors[:,:k]
    comp=np.dot(pcs.T,standardized)
    rec=np.dot(pcs,comp)

    plt.imshow(rec,cmap='gray')
    plt.title(f"PCA Reconstruction k={k}")
    plt.axis("off")
    plt.show()
```
